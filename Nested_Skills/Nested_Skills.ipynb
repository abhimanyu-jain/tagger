```python
# Create list of skills which have no parent (Forest). Depth set to 0
# For each skill, if parent is in Forest, assign to them as children. Add them to depth_1 dictionary
# Then add to depth 2 dictionary and assign as children
```

```python
from pymongo import MongoClient
from bson.objectid import ObjectId
client = MongoClient('127.0.0.1:27017')
db = client.the_incircle_dev_new
print(db.list_collection_names())
```

```python
roots = {}

for skill in db.skills.find({'parent': {"$exists": False}}):
    #print(skill['name'])
    roots[skill['_id']] = True
    #reset(skill, 0, skill)
    
print(roots)
```

```python
non_roots = []

for skill in db.skills.find({'parent': {"$exists": True}}):
    #print(skill['name'])
    non_roots.append(skill)
    #reset(skill, 0, skill)
    
print(len(non_roots))
```

```python
def reset(skill, depth, root):
    depth  = depth + 1
    for sk in db.skills.find():
        #print(sk['parent'])
        parent = db.skills.find_one({'_id' : ObjectId(sk['parent'])})
        if(depth > 1):
            print('parent : ', parent['name'])
            print('child : ', sk['name'])
        
                
```

```python
def getRoot(skill):
    duplicates = {}
    #name = skill['name']
    #parent = db.skills.find_one({'_id' : ObjectId(skill['parent'])})
    #parent_id = parent['_id']
    #print(parent_id)
    print("\n\n\nNew cycle\n")
    print(skill['name'])
    sk = skill
    while(sk['_id'] not in roots):
        if(sk['_id'] in duplicates):
            print('Cycle Detected!!!')
            break
        else:
            duplicates[sk['_id']] = True
        
        if(sk['_id'] == sk['parent']):
            break
        sk = db.skills.find_one({'_id' : ObjectId(sk['parent'])})
        print(sk['name'])
    root = sk['_id']
    return root
    
```

```python
def replaceParent(sk_id, parent):
    db.skills.update_one
    (
        {
            '_id': ObjectId(sk_id)
        }, 

        {
          "$set": { 
           'parent': ObjectId(parent)
          }

        }
    )
```

```python
for skill in non_roots:
    root = getRoot(skill)
    #print(root)
    replaceParent(skill['_id'], root)
```
