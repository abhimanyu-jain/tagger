```python
def print_dictionary(dic):
    for keys,values in dic.items():
        print(keys)
        print(values)
```

```python
skill_names = np.array([])
counts = np.array([])
userprofile_cursor.rewind()

#Iterate over every skill
for userprofile in userprofile_cursor:
    
    ##Obtain list of skills for each userprofile
    userskills = userprofile['skills']
    num_skills = len(userskills)
    
    ##Iterate over list of skills for each userprofile
    for i in range(num_skills):
        ##Find the name of each skill
        skill = skills.find_one({'_id' : userskills[i]})
        
        ## Check for if skill id is not found in 
        if (skill == None):
            continue
        
        skill_name = skill['name']

        if skill_name in skillDict:
            count = skillDict[skill_name]
            count = count + 1
            skillDict[skill_name] = count
        else:
            skillDict[skill_name] = 1
            
        #print_dictionary(skillDict)
        #print('\n\n\n')
            
                
    #data = {'Skill': skill_names, 'UserProfile Occurence Counts': counts}
    #df = pd.DataFrame.from_dict(data)
```
