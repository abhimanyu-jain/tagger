```python
# importing the module
import json
```

```python
# reading the data from the file
with open('word_counts.json') as f:
    data = f.read()
```

```python
# reconstructing the data as a dictionary
js = json.loads(data)
  
print("Data type after reconstruction : ", type(js))
print(js)
```

```python
#Converting to dataframe

import pandas as pd

df = pd.DataFrame(list(js.items()), columns=['Word', 'Count'])

df
```

```python
# Sorting by count

df = df.sort_values('Count', ascending=False)
df
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(df)
```

```python
import openpyxl
df.to_excel("skill_related_word_counts_sorted.xlsx")  
```

```python
# Using pyspellchecker to check spellings of all words
from spellchecker import SpellChecker
```

```python
spell = SpellChecker()
```

```python
df['Correctly spelled'] = df['Word'].apply(spell.correction)
```

```python
df
```

```python
df.to_excel("skill_related_word_counts_sorted_spelling_corrected.xlsx") 
```

```python
#Converting all words to lowercase
df['lowercase'] = df['Correctly spelled'].str.lower() 
```

```python
df
```

```python
# Find if the words are nouns or not
# https://stackoverflow.com/questions/28033882/determining-whether-a-word-is-a-noun-or-not

from nltk.corpus import wordnet as wn
nouns = {x.name().split('.', 1)[0] for x in wn.all_synsets('n')}
```

```python
def isNoun(str):
    if str in nouns:
        return True
    return False
```

```python
df['Noun'] = df['lowercase'].apply(isNoun)
```

```python
df
```

```python
df.to_excel("noun_check.xlsx") 
```
